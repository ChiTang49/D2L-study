### 节省内存

````python
X = torch.arange(12).reshape(3, 4)
Y = torch.zeros(3, 4)

before = id(Y) # id类似C++的指针，指向Y的内存地址
Y = X + Y
after = id(Y)
before == after
````

运行一些操作可能会导致为新结果分配内存 <br>
`Y = X + Y` 后，Y的内存地址会改变 <br>
这会造成内存的浪费<br>
使用 `X[:] = X + Y` 或 `X += Y` 可以减少操作的内存开销

````python
Z = torch.zeros_like(Y)
print('id(Z):', id(Z))
Z[:] = X + Y
print('id(Z):', id(Z))
````

- `Z[:]` 选中 `Z` 的全部元素，但仍然是原来的 `Z` 
- `Z[:] = X + Y`: 计算 `X + Y` ，然后把结果复制到 `Z` 原来的内存空间里。

但是依然会分配新的内存空间（`X + Y` 会被计算出来, 然后复制到 `Z` 原来的内存空间里）

````python
torch.add(X, Y, out=Z)
Z
````

### 范数（torch.norm）
用于衡量一个张量的大小或长度,即从原点到该点的距离。<br>
**深度学习中，通常用于最大化概率/最小化预测与目标之间的距离等**

#### L1 范数
所有元素绝对值之和，也叫曼哈顿距离。
$$
\| \mathbf{x} \|_1 = \sum_{i=1}^{n} |x_i|
$$
```python
torch.norm(x, p=1)
```

````python
x = torch.tensor([2, 3, 4], dtype=torch.float32)
torch.norm(x, p=1)
````

注意范数计算需要浮点数类型，否则会报错。

#### L2 范数
向量各元素平方和再开根号，也叫欧几里得距离。
$$
\| \mathbf{x} \|_2 = \sqrt{\sum_{i=1}^{n} x_i^2}
$$
```python
torch.norm(x) # 默认计算 L2 范数
torch.norm(x, p=2)
```

````python
x = torch.tensor([2, 3, 4], dtype=torch.float32)
torch.norm(x, p=2)
````

#### F范数
就是 L2 范数在矩阵上的推广。
$$
\| \mathbf{A} \|_F = \sqrt{\sum_{i=1}^{m} \sum_{j=1}^{n} a_{ij}^2}
$$
```python
torch.norm(A)
```

````python
x = torch.arange(10, dtype=torch.float32).reshape(2, 5)
torch.norm(x, p=2)
````

## 2.5 自动微分

### 求导

计算 $y = 2x^{T}x$ 的关于列向量 `x` 求导

````python
x = torch.arange(4.0, requires_grad=True)
x, x.grad
````

````python
y = 2 * torch.dot(x, x)
y
````

````python
y.backward()
x.grad
````

`y.backward()` 是反向传播，它会自动计算 `y` 对所有 `requires_grad=True` 的变量的梯度，结果存在 `.grad` 里。

### 求偏导

只计算 $z$ 对 $x$ 的偏导，将 $y$ 的计算图切断（视为常数），只保留 $z$ 的计算图。

$$
y = x^2 \\
z = x^3 \\
令 u = y = x^2，则 u = u * x。\\
$$

````python
x.grad.zero_()     # 清空之前的梯度
y = x * x          # y = x²，保留了计算图
u = y.detach()     # u 的值等于 y，但切断了计算图（PyTorch 把 u 当常数）
z = u * x          # z = u · x，u 被当作常数
````

````python
z.sum().backward()  # 反向传播
x.grad == u
````

故 `y` 对 `x` 的偏导为 $u$。

另一种方法：直接在 `y` 上调用反向传播

````python
x.grad.zero_()
y.sum().backward()
x.grad == 2 * x
````
