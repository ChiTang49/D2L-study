### 节省内存

````python
X = torch.arange(12).reshape(3, 4)
Y = torch.zeros(3, 4)

before = id(Y) # id类似C++的指针，指向Y的内存地址
Y = X + Y
after = id(Y)
before == after
````

运行一些操作可能会导致为新结果分配内存 <br>
`Y = X + Y` 后，Y的内存地址会改变 <br>
这会造成内存的浪费<br>
使用 `X[:] = X + Y` 或 `X += Y` 可以减少操作的内存开销

````python
Z = torch.zeros_like(Y)
print('id(Z):', id(Z))
Z[:] = X + Y
print('id(Z):', id(Z))
````

- `Z[:]` 选中 `Z` 的全部元素，但仍然是原来的 `Z` 
- `Z[:] = X + Y`: 计算 `X + Y` ，然后把结果复制到 `Z` 原来的内存空间里。

但是依然会分配新的内存空间（`X + Y` 会被计算出来, 然后复制到 `Z` 原来的内存空间里）

````python
torch.add(X, Y, out=Z)
Z
````

### 处理缺失值
`NAN` 表示缺失值
`iloc` 方法用于根据索引选择数据

````python
inputs, outputs = data.iloc[:, 0:2], data.iloc[:, 2]
inputs = inputs.fillna(inputs.mean(numeric_only=True)) # 只对数值型NaN填充均值
inputs
````

`get_dummies` 方法用于将分类变量转换为独热编码<br>
把 Alley 展开成多列：

````python
inputs = pd.get_dummies(inputs, dummy_na=True)
print(inputs)
````
